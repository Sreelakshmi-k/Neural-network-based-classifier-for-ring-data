```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
N = 1000
```

```python
theta = 2*np.pi*np.random.rand(N,1);
```

```python
r1=5+np.random.rand(N,1)
x1=r1*np.cos(theta)
y1=r1*np.sin(theta)
#np.shape(r1)
```

```python
r2=10+np.random.rand(N,1)
x2 = r2*np.cos(theta);
y2 = r2*np.sin(theta);
```

```python
plt.title('Ring data')

plt.plot(x1, y1,'.',label='Class 1')
plt.plot(x2,y2,'.',label='Class2')
plt.legend()
plt.show()
```

```python
bias = np.ones((2*N,1))
```

```python
np.shape(bias)
```

```python
Y1 = np.hstack((x1,y1))
Y2=np.hstack((x2,y2))
s=np.vstack((Y1,Y2))
X=np.hstack((s,bias))
np.shape(X)
```

```python
d1=np.zeros(1000)
d2=np.ones(1000)
d=np.hstack((d1,d2))
np.shape(d)
d
```

```python
from keras.models import Sequential
from keras. layers import Dense

model = Sequential()
model.add(Dense(32, input_dim=3, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='mean_squared_error', optimizer = 'sgd', metrics = ['accuracy'])
model.fit(X,d,epochs=1000)
model.summary()
```

```python
y_pred = model.predict(X)
score = model.evaluate(X,d)
```

```python
accuracy = score[1]*100
print(accuracy)
```
